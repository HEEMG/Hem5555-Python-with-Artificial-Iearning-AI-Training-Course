```python
# Image annotation
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('sun.jpg')
image1 = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray = cv2.resize(image1,(50,50))
fig = plt.figure(figsize= (20,20))
ax = fig.add_subplot(111)
r = 0
for rows in gray:
    c = 0
    for pixel in rows:
        ax.annotate(pixel,(r,c))
        c = c+1
    r = r+1
ax.imshow(gray)
plt.show()
```

```python
# image masking
## image masking
import cv2
import numpy as np
import matplotlib.pyplot as plt
# read image
fg = cv2.imread('sun.jpg', 1) # 1 reads as BGR 0 reads as Grayscale
fg = cv2.resize(fg, (425, 425))


# Color changing
rgb_fg = cv2.cvtColor(fg, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_fg)
plt.title('RGB image')
plt.show()


# lets read a garden image
garden = cv2.imread('garden.jpg', 1)

#reshape garden to shape of flag
# print(garden.shape)

garden = cv2.resize(garden, (425, 425))

rgb_garden = cv2.cvtColor(garden, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_garden)
plt.show()


# convert both flower and garden into grayscale

gray_fg = cv2.cvtColor(rgb_fg, cv2.COLOR_BGR2GRAY)
gray_garden = cv2.cvtColor(rgb_garden, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_fg)
plt.show()

plt.imshow(gray_garden)
plt.show()


# create a mask of flag

lv = np.array([0, 0, 0])
hv = np.array([220, 220, 255])

mask = cv2.inRange(fg, lv, hv)
plt.imshow(mask)
plt.show()


masked_img = rgb_fg.copy()
masked_img[mask != 255] = [0, 0, 0]
plt.imshow(masked_img)
plt.show()

bg_cpy = rgb_garden.copy()
bg_cpy[mask == 255] = [0, 0, 0]

plt.imshow(bg_cpy)
plt.show()

final = bg_cpy + masked_img
plt.imshow(final)
plt.show()
```
