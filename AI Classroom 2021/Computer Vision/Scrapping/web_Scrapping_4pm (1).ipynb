```
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
url
```

```
response = requests.get(url)
data = response.text
data = BeautifulSoup(data,'lxml')
data
```

```
final = []
tag = data.table
clear = tag.find_all('tr')
list1 = clear[1:-3]
for i in list1:
  my_data = i.get_text()
  final.append(my_data)
final
```

```
final_data = []
for i in final:
  my_datas = i.split('\n')[1:-1]
  final_data.append(my_datas)

final_data
```

```
from bs4 import BeautifulSoup
import requests
final_data = []
n = int(input("Enter how many page = "))
for i in range(1,n+1):
  url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/', 'html.parser')

  response = requests.get(url)
  data = response.text
  data = BeautifulSoup(data,'lxml')

  final = []
  tag = data.table
  clear = tag.find_all('tr')
  list1 = clear[1:-3]
  for i in list1:
    my_data = i.get_text()
    final.append(my_data)
    
  for i in final:
    my_datas = i.split('\n')[1:-1]
    final_data.append(my_datas)

final_data
```

```
import csv   
fields=final_data
with open('class4pm.csv', 'w') as f:
  x = csv.writer(f)
  for i in fields:
      x.writerow(i)

```
