```
# all 4 authentication keys to access twitter API
# to connect as OAth handler or jump serever / twitters  proxy server
consumer_key = ""
consumer_sec = ""


# from proxy server we need to connect
access_token = ""
access_token_sec = ""
```

```
# connect to jump server of twitter #authenticating the secret code - that I am a valid user of a development account of twitter 
auth = tweepy.OAuthHandler(consumer_key,consumer_sec) #.OAuthHandler is a function that helps to authenticate the consumer key and consumer secret codes above. 
                                                      #We only need consumer key and consumer secret for the task at hand. Access key and access secret are required for packages such as 
                                                      # web application, scrapping in web application. Consumer key & consumer secret are mainly used for sentiment analysis, twitter scrapping and tweepy packages.
```

```
# now we can connect from jump server to web server of twitter. in 3rd party API, we reach the twitter server from out network 
auth.set_access_token(access_token,access_token_sec) #to connect to the servers #when we write auth. it is an object. An object is used to command a certain task. In this case, we are commanding to set access token. 
                                                     # .set_access_token is a predefined function of the tweepy package. 
                                                     
                                                  
```

```
#connect to API storage server of twitter
api_connect=tweepy.API(auth) #.API is a predefined function of the tweepy package 
```

```
hashtag= input("What do you want to search about: .") #input function to create user defined variables 
number= input("Please enter the number of tweets you want download: ")
tweet_data=api_connect.search(hashtag,count=number, lang='en')#.search too is a function within the tweepy package, perhaps present in others too
```

```
my_tweet = set()
for tweet in tweet_data:
 my_tweet.add(tweet.text)#Place all tweets which are in string format into a list called 'my_tweet' 
#Add a header called 'Tweets' in order to create a column of tweets in a csv file 

final_my_tweet = ['Tweets']+list(my_tweet)
final_my_tweet
```

```
b = open('covid_tweets.csv', 'a') #open() is a function for file handling #my_tweets.csv is an argument that creates a csv file #'w' is a mode to overwrite on the created file 
for i in final_my_tweet:
  b.write(f"{i}\n") #.write is a function that writes each and every string or element from the list on to the created file #(f"{i}\n) is an argument for string formatting #'f' is a keyword 
                    # used for string formatting #{i}: if you want to format a variable within a string, you must use curly brackets. 
                    #here, 'i' is a variable that denotes each element in the list. #\n pushes to go to the next line in the csv file after each element is formatted and written. 
b.close()           #b.close() closes the opened file. 
```

```
data = df
new_data = data.drop_duplicates(subset="Tweets") #to drop retweets and repeated rows #subset is a keyword in pandas with column value as "Tweets" to indicate the column with duplicate text
new_data.to_csv('/content/filtered_data.csv')
new_data
```

```
# Text pre-processing 
def get_normalize(text):
    text = re.sub(r'\&\w*;','',text)
    text = re.sub('@[^\s]+','',text)
    text = re.sub(r'\$\w*','',text)
    text = text.lower()
    text = re.sub("[^a-zA-Z]", " ", text) 
    text = re.sub(r'https?:\/\/.*\/\w*','',text)
    text = re.sub(r'#\w*','',text)
    text = re.sub(r'[' + string.punctuation.replace('@','') + ']+',' ',text)
    text = re.sub(r'\b\w{1,2}\b','',text)
    text = re.sub(r'\s\s+',' ',text)
    text = text.lstrip(' ')
    text = ''.join(c for c in text if c <= '\uFFFF')
    return text
j = 0
for i in new_data['Tweets']:
  new_data.loc[j]=get_normalize(str(i))
  j = j+1
new_data
```
